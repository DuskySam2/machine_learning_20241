# K Means - Generated data

Isotropic dataset module in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html) \
Train-test splits in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) \
k-Means module in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

```
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```
_, axes = plt.subplots(1, 2, figsize=(13, 5))

X, _ = make_blobs(n_samples=1000, random_state=170)

# Create train (80%) and test (20%) sets and print all shapes
X_train, X_test = train_test_split(X, test_size=0.2, random_state=1)
print(f"Train set shape: X_train={X_train.shape}")
print(f"Test set shape: X_test={X_test.shape}")

axes[0].scatter(X[:, 0], X[:, 1], edgecolors='k', cmap=plt.cm.Paired, c = 'g' )
axes[0].set_title("Train Set/Test Set")
axes[1].scatter(X_train[:, 0], X_train[:, 1], edgecolors='k', cmap=plt.cm.Paired, c = 'r')
axes[1].scatter(X_test[:, 0], X_test[:, 1], edgecolors='k', cmap=plt.cm.Paired, c = 'b')
axes[1].set_title("Train Set/Test Set")

plt.show()
```

```
kmeans = KMeans(n_clusters=3, random_state = 1)

# Train Kmeans
kmeans.fit(X_train)
```

```
# Plot centroids
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
print(centroids, labels.shape)
```

```
# Create predictions
predictions = kmeans.predict(X_test)
print(predictions.shape)
```

```
_, ax = plt.subplots (figsize = (15,5))
colors=['blue','cyan','magenta']

# plot data and found centroids
```

```
_, axes = plt.subplots(1,2, figsize=(10, 4))
colors = ['blue', 'cyan', 'magenta']

# Plot data and found centroids
for i in range(3):
    axes[0].scatter(X_test[predictions == i][:, 0], X_test[predictions == i][:, 1], color='blue', label=f'Cluster {i + 1}', alpha = 0.5)


# Plot centroids
axes[0].scatter(centroids[:, 0], centroids[:, 1], color='red', marker='*', s=200, label='Centroids')

axes[0].set_title('K-means Clustering')
axes[0].set_xlabel('Feature 1')
axes[0].set_ylabel('Feature 2')
axes[0].legend()
axes[1].scatter(X_test[:,0], X_test[:,1], c=predictions, edgecolor = 'k', cmap = 'Paired')

plt.show()
```

Naive Bayes documentation [Aquí](https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.GaussianNB.html#sklearn.naive_bayes.GaussianNB)

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris # Iris dataset
```

```
# Load the iris dataset
iris = load_iris()

# Select 2 features
iris_x = iris.data[:, :2]
feature_names = iris.feature_names

# Get labels
iris_y = iris.target

# Print shapes
print(f"Features shape: {iris_x.shape}")
print(f"Labels shape: {iris_y.shape}")
print(f"Labels shape: {feature_names}")

_, ax = plt.subplots (figsize=(5,4))
ax.scatter(iris_x[:, 0], iris_x[:, 1], c = iris_y, cmap = 'Paired')
ax.set_xlabel (feature_names[0])
ax.set_ylabel(feature_names[1])
```

```
# Instance kNN object
naive_bayes = GaussianNB()

# Training
naive_bayes.fit( iris_x , iris_y )
```

```
## Plot decision boundaries
def plot_decision_boundaries(X, y, clf):

    # Definir fronteras
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
    h = .01  # step size in the mesh
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = naive_bayes.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap='Paired')

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='Paired', edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title('Decision Boundary of Gaussian Naive Bayes')

    plt.show()

plot_decision_boundaries(iris_x, iris_y, naive_bayes)
```
